# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
import requests
import pprint
btc = requests.get("https://api.bithumb.com/public/ticker/btc").json()['data']
pprint.pprint(btc)
````

````python
# 아래에 코드를 작성하세요.
difference = int(btc['max_price']) - int(btc['min_price'])
if int(btc['closing_price']) + difference > int(btc['max_price']):
    print('상승장')
else :
    print('하락장')
````

## 모든 가상화폐 시세 판단하기

> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.
>
> 예외 처리에 유의하세요.
--- 
```
예시 출력)
BTC 상승장
ETH 상승장
DASH 상승장
LTC 하락장
ETC 하락장
XRP 상승장
BCH 상승장
...
```


````python
import pprint
currency = requests.get("https://api.bithumb.com/public/ticker/all").json()['data']
pprint.pprint(currency)
````

````python
# 아래에 코드를 작성하세요
up, down, total = 0, 0, 0
for coin in currency.keys() :
    if 'max_price' in currency[coin] :
        total += 1
        difference = float(currency[coin]['max_price']) - float(currency[coin]['min_price'])
        if float(currency[coin]['closing_price']) + difference > float(currency[coin]['max_price']):
            print(f'{coin} : 상승장')
            up += 1
        else :
            print(f'{coin} : 하락장')
            down += 1
if total * 3 / 10 < up :
    print('곧 떡락합니다.')
if total * 1 / 2 < down :
    print('지금 사야합니다.')
````

````python
# 아래에 코드를 작성하세요
up, total = 0, 0
for coin in currency.keys() :
    cc = currency[coin]
    if 'max_price' in cc :
        total += 1
        if float(cc['closing_price']) > float(cc['min_price']):
            up += 1
if total * 3 / 10 < up :
    print('곧 떡락합니다.')
if up < total * 1 / 2 :
    print('지금 사야합니다.')
````

# 영어 이름 출력하기 

> 영어 이름은 가운데 이름을 가지고 있는 경우가 있습니다.
>
> 가운데 이름은 축약해서 나타내는 함수를 작성해보세요.

---
```
예시 입력)
Alice Betty Catherine Davis
```
```
예시 출력)
Alice B. C. Davis
```

````python
# 아래에 코드를 작성하세요.
names = 'Alice Betty Catherine Davis'
listnames = list(names)
for i in range(0, len(listnames)):
    print(listnames[i], end = '')
    if listnames[i] == ' ':
        check = i
        while listnames[check] != ' ' :
            if listnames[check] == None:
                break
            check += 1
        if listnames[check] == None:
                break
        i += 1
        print(f'{listnames[i]}.')
        while listnames[i] != ' ':
            i += 1
````

````python
names = 'Alice Betty Catherine Davis'
name = ''
list_names = names.split()
for nums in range(0, len(list_names)) :
    if not list_names[0] == list_names[-nums] :
        if not list_names[-1] == list_names[-nums] :
            list_names[-nums] = list_names[-nums][0:1] + '.'
for short_name in list_names :
    name += short_name + ' '
print(name)
````

````python
# 아래에 코드를 작성하세요.
import calendar
print(calendar.calendar(1))
cnt = 0
for x in range(1, 13):
    print(f'         {x} 월')
    print(f'Mo Tu We Th Fr Sa Su')
    if x in (1, 3, 5, 7, 8, 10, 12):
        for j in range(0, cnt):
            print('   ', end = '')
        for i in range(1, 32):
            print(f'{i}  ', end = '')
            cnt += 1
            if cnt == 7:
                print(f'')
                cnt = 0
        print('')
    elif x == 2:
        for j in range(0, cnt):
            print('   ', end = '')
        for i in range(1, 29):
            print(f'{i}  ', end = '')
            cnt += 1
            if cnt == 7:
                print(f'')
                cnt = 0
        print('')
    else :
        for j in range(0, cnt):
            print('   ', end = '')
        for i in range(1, 31):
            print(f'{i}  ', end = '')
            cnt += 1
            if cnt == 7:
                print(f'')
                cnt = 0
        print('')
````

````python
def TaeYeonNum(number) :
    result = 0
    for numb in number :
        num = int(numb)
        if num % 2 :
            result *= num
        elif num == 0 :
            continue
        else :
            result += num
    return result
numbers = input()
print(TaeYeonNum(numbers))
````
